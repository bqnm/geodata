```python
def pull_geometry(x):
    geos = eval(x)
    coords = geos['coordinates']
    return {'longitude': coords[0], 'latitude': coords[1]}
    

geometry_dicts = df1.geometry.map(pull_geometry).to_list()
geometry_df = pd.DataFrame(geometry_dicts, index=df1.index)
final_df = pd.concat([df1, geometry_df], axis=1)
```
