```python
table = soup.findAll('table')[0]
```

```python
def get_ratings_info(table):
    info = {}
    rows = table.findAll('tr')
    del rows[0]
    for row in rows:
        columns = row.findAll('td')
        info[columns[0].find('a').contents[0]] = columns[3].find('a').contents[0]
        
    return info
```
